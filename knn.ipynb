```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


csv_df=pd.read_csv('iris.csv')
display(csv_df)
```

```python
# part 1.
# spliting of data set for each given vaariety for both part a and b

## setosa
condition_setosa=csv_df[csv_df['variety']=='Setosa']

# testing data set
condition_setosa_test=condition_setosa[:10]

# training data set
condition_setosa_train=condition_setosa[-40:]

## virginica
condition_virginica=csv_df[csv_df['variety']=='Virginica']

# testing data set
condition_virginica_test=condition_virginica[:10]

# training data set
condition_virginica_train=condition_virginica[-40:]

##versicolor
condition_versicolor=csv_df[csv_df['variety']=='Versicolor']

# testing data set
condition_versicolor_test=condition_versicolor[:10]

# training data set
condition_versicolor_train=condition_versicolor[-40:]


```

```python
# part 2
# ploting data using scatter plot

# create figure size
plt.figure(figsize=(10,10))


# plot of training dataset of setosa
x_setosa=condition_setosa_train['sepal.length']
y_setosa=condition_setosa_train['sepal.width']

plt.scatter(x_setosa,y_setosa,c='red',label='Iris_setosa')

# plot of traning dataset of virginica
x_virginica=condition_virginica_train['sepal.length']
y_virginica=condition_virginica_train['sepal.width']

plt.scatter(x_virginica,y_virginica,c='blue',label='Iris_virginica')

# plot of training dataset of versicolor
x_versicolor=condition_versicolor_train['sepal.length']
y_versicolor=condition_versicolor_train['sepal.width']

plt.scatter(x_versicolor,y_versicolor,c='green',label='Iris_versicolor')

# set labels and titles
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.legend()
plt.show()
```

```python
# part 3
# computing euclidean distance for each sample in test dataset

import math as m

list1=[]

# concat the test datasets
test_set = pd.concat([condition_setosa_test,condition_virginica_test,condition_versicolor_test],axis=0,ignore_index=True)
train_set = pd.concat([condition_setosa_train,condition_virginica_train,condition_versicolor_train],axis=0,ignore_index=True)
display(train_set)
# calculate euclidean distance for each sample
for index, row in test_set.iterrows():

  list2=[]
  sepal_length_testvalue = row['sepal.length']
  sepal_width_testvalue = row['sepal.width']

  for index,row in train_set.iterrows():
    sepal_length_trainvalue = row['sepal.length']
    sepal_width_trainvalue = row['sepal.width']

    euclidean_distance=m.sqrt((sepal_length_testvalue-sepal_length_trainvalue)**2 + (sepal_width_testvalue-sepal_width_trainvalue)**2)
    list2.append(float(euclidean_distance))

  list1.append(list2)

# convert list into data frame
distance_df=pd.DataFrame(list1).transpose()

display(distance_df)

```

```python
# part 4
# Identify the K-nearest neighbors with the smallest Euclidean distances

# k is hyperparameter
# also value of k need to be odd bcoz it will avoid the ambigious condition

k=int(input("enter value of k: "))

# find nearest k neighbors for each sample
nearest = {}

for i in distance_df.columns:
  smallest_values = distance_df[i].nsmallest(k)
  nearest[i] = smallest_values.values

nearest_neighbors = pd.DataFrame(nearest)

# keep track of index of nearest neighbor for each sample
nearest_neighbors_index=distance_df.apply(lambda col: col.nsmallest(k).index)

display(nearest_neighbors)
print(nearest_neighbors_index)
```

```python
# part 5
# majority class among the K-nearest neighbors and assign it as the predicted class for the sample.

list_variety=[]

for i in nearest_neighbors_index:
  temp=[]
  for j in range(k):
    temp.append(train_set.loc[(nearest_neighbors_index.iloc[j,i]),'variety'])
  list_variety.append(temp)

neighbor_class=pd.DataFrame(list_variety).transpose()
display(neighbor_class)

# calculate most repeated value using mode

majority_class = neighbor_class.mode()

print("predicted class for each sample: ")
display(majority_class.T)
```

```python
# part 6
# calculate accuracy of classifier
# we have accuracy of knn as ratio correctly predicted to total predicted

count=0
predicted_values = majority_class.T

for i in range(len(predicted_values[0])):
  if(test_set.loc[i,'variety'] == predicted_values.iloc[i,0]):
    count += 1

model_accuracy = (count/len(predicted_values[0]))*100
print("no of samples predicted correctly: ",count)
print("Accuracy of Model: ",model_accuracy)
```

```python
# now use all attributes in dataset
# calculate euclidean distance for each sample
list3=[]

for index, row in test_set.iterrows():

  list2=[]
  sepal_length_testvalue = row['sepal.length']
  sepal_width_testvalue = row['sepal.width']
  petal_length_testvalue = row['petal.length']
  petal_width_testvalue = row['petal.width']
  for index,row in train_set.iterrows():
    sepal_length_trainvalue = row['sepal.length']
    sepal_width_trainvalue = row['sepal.width']
    petal_length_trainvalue = row['petal.length']
    petal_width_trainvalue = row['petal.width']

    euclidean_distance=m.sqrt((sepal_length_testvalue-sepal_length_trainvalue)**2 + (sepal_width_testvalue-sepal_width_trainvalue)**2 + (petal_length_testvalue-petal_length_trainvalue)**2 + (petal_width_testvalue-petal_width_trainvalue)**2)
    list2.append(float(euclidean_distance))

  list3.append(list2)

distance_df_1=pd.DataFrame(list3).transpose()
display(distance_df_1)
```
